buffering is a temporary space where the data is stored for the temporary basis during file handling before the read and write operation .This increase the efficency of the i/o operation.

In case of writing the data is loaded in the buffer before it is flushed to the file in single write operation.

In case of Reading the large chunk of data from the file is stored in the buffer space and it is read in single read operation. 

Qeustion 7

Steps involved in buffering in python are:-
1. opening the file using open() with buffered parameter set.
2. performing the reading and writing operation using read and write functions
3. managing the buffer using appropiate flush() method.
4. closing the file using file.close()
for example :-

```python
file_path = 'Text.txt'
buffer_size = 100
with open(file_path, 'w', buffering=buffer_size) as f:
    for i in range(1000):
        f.write(f'Line {i}\n')
        if i % 100 == 0:  
            f.flush()

print("Data written with full buffering.")
```

#question 27
this are specific exception that are specific to a program or application .In order to handel some specific exception in more descriptive way we can create a custom exception , by defining our own exception class.

To create custom exception class we need to create a class that inherit the exception class or its subclass.

```python
#question 28
class Custom(Exception):
    def __init__(self,message):
        super().__init__(message)
    def __str__(self):
        return f"the error occured {self.args[0]}"
        
def operation(num):
        if num<0:
            raise Custom("Negative Value Error")
        num = num - 1
        return num
try:
    result = operation(-5)
    print(f"the result is {result}")
except Custom as e:
    print(e)
```

```python
#questiom 29
class CustomError(Exception):
    def __init__(self,message,error_code):
        self.error_code = error_code
        super().__init__(message)
    
    def __str__(self):
        return f"Custom Error occured {self.args[0]} error code : {self.error_code}"
def operation(value):
    if value%2 == 0:
        raise CustomError("value is even",1201)
    return value
try:
    result = operation(64)
    print(f"the result is {result}")
except CustomError as e:
    print(e)
```

```python
#question 30
class NegativeError(Exception):
    def __init__(self,message):
        super().__init__(message)
    def __str__(self):
        return f"the value cannot be less than zero: {self.args[0]}"
def neg(value):
    if value < 0:
        raise NegativeError("NeagtiveError")
    return value
try:
    result = neg(-2)
    print(f"the result is {result}")
except NegativeError as e:
    print(e)
    
```

#question 31
try,except,else,finally are used in exception handling.Each bloxk has its own significance .

Try Block :- the try block contains the code that might raise an exception. This is the block where we place the code that we want to monitor for errors.

Except:-The except block catches and handles exceptions that are raised in the try block. we can specify the type of exception we want to catch and handle.

Else:-The else block contains code that should be executed if no exceptions were raised in the try block. This is useful for code that should only run if the try block was successful.

Finally:-The finally block contains code that should be executed no matter what, whether an exception was raised or not. This is typically used for cleanup actions like closing files or releasing resources.

Multithreading is a concurrent execution model that allows multiple threads to exist within the context of a single process. These threads share the process's resources but are able to execute independently.

```python
#question 34
import threading
import time

def square(num):
    print(f"the sqaure of {num} is",num**2)
t1 = threading.Thread(target = square,args =(5,))
t2 = threading.Thread(target = square , args = (6,))
start = time.perf_counter()
t1.start()
t2.start()

t1.join()
t2.join()
end = time.perf_counter()
print(f"the total excution time {end - start}")
```

#question 40
Difference between multithreading and multiprocessing are:-
1. Multithreading involves running multiple threads within a single process. Each thread shares the same memory space and can access the same variables whereas in Multiprocessing involves running multiple processes, each with its own memory space. Each process runs independently of the others.
2. multithreading is useful for I/O-bound tasks where the program spends a lot of time waiting for external events (e.g., file I/O, network communication).
Multiprocessing is useful for CPU-bound tasks where the program performs intensive computations that can benefit from parallel execution on multiple CPU cores.
3. Multithreading : the Global Interpreter Lock (GIL) prevents multiple native threads from executing Python bytecodes at once.
   multiprocessing:Each process has its own Python interpreter and memory space, so the GIL does not affect multiprocessing. Multiple processes can run simultaneously on different CPU cores.
4. multihthreading : Threads within the same process share the same memory space, which makes data sharing easy but can lead to complex synchronization issues and potential data corruption.
   multiprocessing :Each process has its own separate memory space. Data sharing between processes requires explicit inter-process communication (IPC) mechanisms like pipes, queues, or shared memory.

```python
#question 41

import multiprocessing 
import time

def square(num,output_que):
    message = f"the square of the {num} is {num**2}\n"
    time.sleep(2)
    message+=f"process for {num} is finished"
    output_que.put(message)
    

if __name__ == '__main__':
    start = time.perf_counter()
    processes = []
    output_que = multiprocessing.Queue()
    for i in range(5):
        p = multiprocessing.Process(target = square,args = (i,output_que))
        p.start()
        processes.append(p)
    for process in processes:
        process.join()

    while not output_que.empty():
        print(output_que.get())
        
    end = time.perf_counter()
    print("all the processes are executed.")
    print(f"Execution time is {end - start}")
```

```python
#question 42
import multiprocessing 
import time

def square(num):
    return num*num
start = time.perf_counter()
if __name__ == '__main__':
    with multiprocessing.Pool(processes = 4) as pool:
        result = pool.map(square,range(5))
        print(result)
    end = time.perf_counter()
    print(f"Total execution time {end -start}")
```
